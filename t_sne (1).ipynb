```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import plotly.express as px
```

```python
mnist = load_digits()
target = mnist.target
mnist = mnist.data
```

```python
tsne = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(mnist)
pca = PCA(n_components=2).fit_transform(mnist)
```

```python
px.scatter(x=tsne[:, 0], y=tsne[:, 1], color=target)
```

```python
px.scatter(x=pca[:, 0], y=pca[:, 1], color=target)
```
